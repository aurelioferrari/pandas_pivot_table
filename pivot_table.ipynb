```python
import pandas as pd
```

```python
carros = [7, 4, 3, 2, 8]
dias = pd.date_range(start='20190101', end='20190101', periods=5)
vendedor = ['Geroge', 'Vagner', 'Pedro', 'Vagner', 'Geroge']
df = pd.DataFrame({'Vendas': carros, 'Data': dias, 'Vendedor': vendedor})
```

```python
from pandas.core.reshape.pivot import AggFuncType
pd.pivot_table(df, index='Data', columns='Vendedor', values='Vendas', aggfunc='sum')
```
